```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('data.csv')
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df, columns=['fuel', 'transmission', 'seller_type'])
```

```python
df.head()
```

```python
df['fuel_CNG']=df['fuel_CNG'].astype(int)
df['fuel_Diesel']=df['fuel_Diesel'].astype(int)
df['fuel_Electric']=df['fuel_Electric'].astype(int)
df['fuel_LPG']=df['fuel_LPG'].astype(int)
df['fuel_Petrol']=df['fuel_Petrol'].astype(int)
df['transmission_Automatic']=df['transmission_Automatic'].astype(int)
df['transmission_Manual']=df['transmission_Manual'].astype(int)
df['seller_type_Individual']=df['seller_type_Individual'].astype(int)
df['seller_type_Dealer']=df['seller_type_Dealer'].astype(int)
df['seller_type_Trustmark Dealer']=df['seller_type_Trustmark Dealer'].astype(int)
```

```python
df
```

```python
df['owner']=df['owner'].map({
    'First Owner': 1,
    'Second Owner': 2,
    'Third Owner': 3,
    'Fourth & Above Owner': 4
})
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['km_driven'] = scaler.fit_transform(df[['km_driven']])
```

```python
df
```

```python
df['owner']
```

```python
x= df.drop(['name', 'selling_price'], axis=1)
y= df['selling_price']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.info()
```

```python
model=LinearRegression()
```

```python
# Drop rows with NaN values in x_train and y_train
x_train_clean = x_train.dropna()
y_train_clean = y_train.loc[x_train_clean.index]

model.fit(x_train_clean, y_train_clean)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Drop rows with NaN values in x_test and align y_test
x_test_clean = x_test.dropna()
y_test_clean = y_test.loc[x_test_clean.index]

y_pred = model.predict(x_test_clean)
mae = mean_absolute_error(y_test_clean, y_pred)
mse = mean_squared_error(y_test_clean, y_pred)
r2 = r2_score(y_test_clean, y_pred)
print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
